## 💾 Salva immagini generate

```python
# Salva le immagini generate nella cartella /kaggle/working/hachi-gallery
# Poi vai su: Kaggle → Output → salva come dataset 'hachi-gallery'
# e aggiungilo come Input nelle prossime sessioni per ritrovare le immagini

import shutil, os
from pathlib import Path

output_dir = os.environ.get('output_dir', '/root/A1111/outputs')
gallery_dst = Path('/kaggle/working/hachi-gallery')
gallery_dst.mkdir(parents=True, exist_ok=True)

import glob
images = glob.glob(f'{output_dir}/**/*.png', recursive=True) + \
         glob.glob(f'{output_dir}/**/*.jpg', recursive=True) + \
         glob.glob(f'{output_dir}/**/*.webp', recursive=True)

copied = 0
for img in images:
    dst = gallery_dst / Path(img).name
    if not dst.exists():
        shutil.copy2(img, dst)
        copied += 1

print(f'✅ {copied} nuove immagini salvate in {gallery_dst}')
print(f'📁 Totale immagini nella galleria: {len(list(gallery_dst.iterdir()))}')
print()
print('👉 Ora vai su Kaggle → scheda Output → salva come dataset "hachi-gallery"')
```
